```python
#ex 3
# formula IMC = peso / altura **2
def IMC(peso, altura):
  return peso / altura **2

def classificaIMC(peso, altura):
  valorIMC = IMC(peso, altura)
  if (valorIMC >30):
    return 'obeso'
  elif (valorIMC > 25):
    return 'sobrepeso'
  elif (valorIMC > 18.5):
    return 'Eutotrofico'
  else:
    return 'Magro'
print (classificaIMC(60, 1.95))
```
